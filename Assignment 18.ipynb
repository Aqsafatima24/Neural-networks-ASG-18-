```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # Hidden layer 1
    Dense(32, activation='relu'),  # Hidden layer 2
    Dense(len(np.unique(y_encoded)), activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

#### Training the Model

### Hyperparameter Tuning

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

# Define model-building function
def create_model(learning_rate=0.01, activation='relu'):
    model = Sequential([
        Dense(64, activation=activation, input_shape=(X_train.shape[1],)),
        Dense(32, activation=activation),
        Dense(len(np.unique(y_train)), activation='softmax')  # Corrected to use `y_train`
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Wrap model with KerasClassifier
model = KerasClassifier(model=create_model, epochs=20, batch_size=32, verbose=0)

# Define hyperparameter grid with model__ prefix
param_grid = {
    'model__learning_rate': [0.001, 0.01, 0.1],
    'model__activation': ['relu', 'tanh'],
    'batch_size': [16, 32, 64],
    'epochs': [10, 20]  # Optional: add epochs to param grid if you want to tune it
}

# Set up GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train, y_train)

# View the best hyperparameters
print(f"Best Parameters: {grid_result.best_params_}")
print(f"Best Score: {grid_result.best_score_}")
```

```python
print("Best Hyperparameters:", grid_result.best_params_)
best_model = grid_result.best_estimator_
```

### Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predictions on test set
y_pred_best = best_model.predict(X_test)
```

```python
# Metrics
accuracy = accuracy_score(y_test, y_pred_best)
precision = precision_score(y_test, y_pred_best, average='weighted')
recall = recall_score(y_test, y_pred_best, average='weighted')
f1 = f1_score(y_test, y_pred_best, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
